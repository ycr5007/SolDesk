```python
df_krx = pd.read_csv('krx.csv')
df_krx.head()
```

#### 종목명으로 코드(Symbol) 찾아오기

```python
# Name 이 하이브 인 row 찾기
df_krx.query("Name == '하이브'")['Symbol'].values[0]
```

```python
def item_code_by_item_name(item_name):
    """
    종목명을 받아 종목코드 반환
    """
    symbol_code = df_krx.query(f"Name == '{item_name}'")['Symbol']
    if len(symbol_code) > 0:
        return symbol_code.values[0]
    else:
        return False
```

```python
def find_item_list(find_name, year=2020):
    """
    종목명, 연도를 받아 리스트를 반환 ( 입력연도가 없을 시 2020 으로 Default 설정 )
    """
    try:
        return fdr.DataReader(item_code_by_item_name(find_name), str(year))
    except:
        return False
```

```python
stock_list = [
    ['삼성전자', '005930'],
    ['SK하이닉스', '000660'],
    ['현대차', '005380'],
    ['셀트리온', '068270'],
    ['LG화학', '051910'],
    ['POSCO', '005490'],
    ['삼성물산', '028260'],
    ['NAVER', '035420']
]

df_list = [fdr.DataReader(code, '2021-01-01', '2021-12-31')['Close'] for name, code in stock_list]
```

```python
len(df_list)
```

```python
df = pd.concat(df_list, axis=1)
df
```

```python
df.columns = [name for name, code in stock_list]
```

```python
df.head(3)
```

```python
df.plot()
```

```python
df[['삼성전자', 'LG화학']].plot(secondary_y='LG화학')
```

#### 기간수익률

```python
df2 = df[['삼성전자', 'LG화학']]
df2
```

```python
df_plot = df2 / df2.iloc[0] - 1
```

```python
df_plot.plot()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))
df[['삼성전자', 'LG화학']].plot(secondary_y='LG화학', ax=axes[0], title='종가')
df_plot.plot(ax=axes[1], title='수익률')
```
