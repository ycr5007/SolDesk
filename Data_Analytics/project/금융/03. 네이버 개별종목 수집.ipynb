```python
def get_day_list(code, page):
    """
    일자별 시세 페이지 데이터 수집
    """
    url = f"https://finance.naver.com/item/sise_day.naver?code={code}&page={page}"
    res = requests.get(url, headers=headers)
    soup = BeautifulSoup(res.text, 'lxml')
    table = soup.select('table')
    temp = pd.read_html(str(table))[0].dropna()
    return temp
```

```python
get_day_list(item_code, 12)
```

```python
import time
```

```python
item_list = []
for p in range(1, 13):
    table = get_day_list(item_code, p)
    item_list.append(table)
    time.sleep(0.3)
```

```python
df = pd.concat(item_list)
```

```python
# 종목코드, 종목명 컬럼 생성
df['종목코드'] = item_code
df['종목명'] = item_name
df.head()
```

```python
# 컬럼 순서 변경
cols = ['종목코드', '종목명', '날짜', '종가', '전일비', '시가', '고가', '저가', '거래량']
df = df[cols]
```

```python
# 함수작성
def get_item_list(item_code, item_name):
    item_list = []
    
    for p in range(1, 13):
        table = get_day_list(item_code, p)
        item_list.append(table)
        time.sleep(0.3)
        
    df = pd.concat(item_list)
    df['종목코드'] = item_code
    df['종목명'] = item_name
    
    cols = ['종목코드', '종목명', '날짜', '종가', '전일비', '시가', '고가', '저가', '거래량']
    df = df[cols]
    
    return df
```
