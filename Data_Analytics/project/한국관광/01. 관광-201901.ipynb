#### 기준년월 컬럼 생성

```python
kto_201901['기준년월'] = '2019-01'
```

#### 아시아주, 미주, 구주, 대양주, 아프리카주, 기타대륙, 교포소계 국적 제거

```python
kto_201901_country = kto_201901[~kto_201901['국적'].isin(['아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙', '교포소계'])]
```

```python
kto_201901_country.head()
```

```python
kto_201901_country = kto_201901_country.reset_index().drop(columns='index')
```

```python
kto_201901_country.index
```

#### 대륙칼럼 추가

```python
continents = ['아시아'] * 25 + ['아메리카'] * 5 + ['유럽'] * 23 + ['오세아니아'] * 3 + ['아프리카'] * 2 + ['기타대륙'] + ['교포']
```

```python
kto_201901_country['대륙'] = continents
```

#### 국적별 관광객 비율
- 관광객비율 = 관광 / 계 * 100 ( 소수점 1 자리 )

```python
kto_201901_country['관광객비율'] = round(kto_201901_country['관광'] / kto_201901_country['계'] * 100, 1)
```

#### 관광객 비율 상하위 5개국 추출

```python
kto_201901_country.sort_values('관광객비율', ascending=False).head(5)
```

```python
kto_201901_country.sort_values('관광객비율').head(5)
```

#### 대륙별 관광객 비율 평균

```python
kto_201901_country.groupby('대륙')[['관광객비율']].mean().sort_values('관광객비율', ascending=False)
```

#### 기준년월별로 전체 외국인 관광객 대비 국적별 관광객 비율
- 전체비율(%) = 관광 / 관광목적의 전체 외국인 수 * 100

```python
sum(kto_201901_country['관광'])
```

```python
kto_201901_country['전체비율'] = round(kto_201901_country['관광'] / sum(kto_201901_country['관광']) * 100, 1)
```
