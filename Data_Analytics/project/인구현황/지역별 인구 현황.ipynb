#### 데이터 전처리

###### 1) 불필요 컬럼 제거

```python
df.drop(columns=['2022년06월_계_총인구수', '2022년06월_계_연령구간인구수'], axis=1, inplace=True)
df.head()
```

```python
df.columns = df.columns.str.replace('2022년06월_계_', '')
df.head()
```

#### 데이터 분석

###### ex1 ) 종로구 평창동 인구 구조
- 종로구 평창동 찾기

```python
df[df['행정구역'].str.contains('평창동')]
```

#### 행정구역 인덱스 설정

```python
df.set_index('행정구역', inplace=True)
df.head()
```

- 인덱스에서 값 찾기

```python
df[df.index.str.contains('평창동')]
```

#### 행정구역 인덱스 해제

```python
df.reset_index(inplace=True)
df.head()
```

```python
dong = df[df['행정구역'].str.contains('평창동')]
dong
```

### 평창동 df 에서 행정구역 컬럼 제거

```python
dong.drop(columns=['행정구역'], inplace=True)
```

#### 행/열 변경 : T, transpose()

```python
data = dong.T
data
```

#### 컬럼명 변경 : 인구수

```python
data.columns = ['인구수']
```

#### 선차트

```python
plt.figure(figsize=(9, 6))
plt.title("평창동 2022년 06월 기준", fontsize=18)
plt.plot(data['인구수'])
```

#### 막대차트

```python
data.plot.bar(rot=45, figsize=(9, 6), title="평창동 2022년 06월 기준")
```

```python
plt.figure(figsize=(9, 6))
plt.title("평창동 2022년 06월 기준", fontsize=18)
plt.barh(width=data['인구수'], y=data.index, color='lightblue', label="인구 수")
plt.legend()
```

## 동네 인구구조 알아보기


#### 1) 사용자 입력

```python
user = input("찾아보고자 하는 동 이름을 입력해주세요\n")
```

#### 2) 사용자 입력값으로 동을 찾아 새로운 데이터 프레임 생성

```python
user_data = df[df['행정구역'].str.contains(user)]
user_data
```

#### 3) 데이터 시각화

```python
user_data.drop(columns=['행정구역'],inplace=True)
```

```python
chart_data = user_data.T
```

```python
chart_data.columns = ['인구수']
```

```python
plt.figure(figsize=(12, 6))
plt.title(f"{user}의 2022년 06월 기준 인구현황")
plt.bar(height=chart_data['인구수'], x=chart_data.index)
```

```python
plt.figure(figsize=(12, 12))
plt.title(f"{user}의 2022년 06월 기준 인구현황", fontsize=18)
plt.pie(chart_data['인구수'], labels=chart_data.index, autopct='%.1f%%')
plt.legend(loc=(1.2, 0.3), title="연령별")
plt.show()
```
