### pandas 차트 작성
* https://pandas.pydata.org/docs/user_guide/visualization.html

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['axes.unicode_minus'] = False # " - " 기호 Unicode 처리
plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
```

```python
df = pd.DataFrame(np.random.randn(1000, 3), columns=list('ABC'))
df = df.cumsum()

df.head()
```

#### 1) 기본 선 그래프 ( 범례 기본 )

```python
df.plot()
```

```python
df['B'].plot()
```

#### 2) 범례 제거

```python
df.plot(legend=False)
```

#### 3) 보조 축 설정

```python
df['A'].plot()
df['B'].plot(secondary_y=True)
```

#### 4) 보조 축 + 축 레이블 지정 

```python
ax = df.plot(secondary_y=['A', 'B'])
ax.set_ylabel('C_scale')
ax.right_ax.set_ylabel('AB_Scale')
```

```python
ax = df.plot(secondary_y=['A', 'B'], mark_right=False) # 범례의 (right) 표시 제거
ax.set_ylabel('C_scale')
ax.right_ax.set_ylabel('AB_Scale')
```

#### 5) subplots

```python
df.plot(subplots=True, figsize=(5,5))
```

```python
df.plot(subplots=True, figsize=(10,10), layout=(2,2), sharex=True, sharey=True)
plt.suptitle('Pandas Subplots')
```
