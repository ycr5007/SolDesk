### area plot ( 면적그래프 )
- NaN 값이 있으면 자동으로 0 이 채워진다
- default 값으로 누적시켜서 작성된다

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=list('abcd'))
df.head()
```

```python
df.plot.area()
```

```python
df.plot.area(stacked=False, grid=True)
```
