## matplotlib (https://matplotlib.org/stable/index)

- 다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['axes.unicode_minus'] = False # " - " 기호 Unicode 처리
plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
plt.rcParams['font.size'] = 15 # 폰트 사이즈
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.title('꺽은선 그래프', fontdict={'size':20})
g = plt.plot(x,y)
```

### 축설정
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html

#### 1. 이름 설정 
><b>1) xlabel(), ylabel()</b>

```python
plt.title('꺽은선 그래프', fontdict={'size':20})

plt.xlabel('X 축')
plt.ylabel('Y 축')

g = plt.plot(x,y)
```

><b>2) 이름 및 다양한 설정</b>

matplotlib.pyplot.xlabel(xlabel, fontdict=None, labelpad=None, *, loc=None, **kwargs)

- fontdict : font 설정을 dict 구조로 설정
- labelpad : default 4.0 설정됨(label과 차트 사이의 간격)
- **kwargs(?)
    - color : rgb설정, 색상명 가능
- loc : 축 이름 위치 설정('left', 'center', 'right'),  y 축 (top, bottom, center)

```python
plt.title('꺽은선 그래프', fontdict={'size':20})

plt.xlabel('X 축', color='red', loc='right')
plt.ylabel('Y 축', color='#00aa00', loc='top')

g = plt.plot(x,y)
```

```python
# 시스템에 저장된 폰트 리스트 확인
import matplotlib.font_manager as fm
fm.fontManager.ttflist

[f.name for f in fm.fontManager.ttflist]
```

```python
fontdict1 = {"family":"serif", "color":"b", "weight":"bold", "size":14}
fontdict2 = {"family":"Consolas", "color":"deeppink", "weight":"bold", "size":16}

plt.title('꺽은선 그래프', fontdict={'size':20})

plt.xlabel('X', color='red', loc='right', fontdict=fontdict1, labelpad=15)
plt.ylabel('Y', color='#00aa00', loc='top', fontdict=fontdict2, labelpad=20)

g = plt.plot(x,y)
```

#### 2. 축 눈금 설정 - xticks(), yticks()

```python
plt.plot(x,y)

plt.xticks([1,2,3], fontsize=15)
plt.yticks([3,6,9,12], fontsize=20)
plt.show()
```
