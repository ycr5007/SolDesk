```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly
import matplotlib.pyplot as plt
```

```python
plt.rcParams['axes.unicode_minus'] = False # " - " 기호 Unicode 처리
plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
plt.rcParams['font.size'] = 15 # 폰트 사이즈
```

### 기본 차트 생성

><b>리스트를 그래프로 나타내기</b>

```python
plt.plot([10,20,30,40])
plt.show()
```

```python
x = [1,2,3,4]
y = [12,43,25,15]

plt.plot(x, y)
```

><b>파이썬 딕셔너리를 그래프로 나타내기</b>

```python
dict = {"data_x" : [1,2,3,4,5], "data_y" : [2,3,5,10,8]}

plt.plot("data_x", "data_y", data=dict)
```

### 타이틀 설정

```python
x = [1,2,3,4]
y = [12,43,25,15]

plt.title('한글 불가 Can English') # 한글 설정시 오류 ( 폰트 설정 필요 )
plt.plot(x, y)
```

### 마이너스 설정

```python
x = [1,2,3,4]
y = [12,-5,25,15]

plt.title('꺽은선 차트')
plt.plot(x, y)
```

```python
plt.rcParams['axes.unicode_minus'] = False # " - " 기호 Unicode 처리
```
