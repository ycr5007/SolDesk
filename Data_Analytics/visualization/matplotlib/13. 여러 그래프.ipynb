```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.rcParams['axes.unicode_minus'] = False # " - " 기호 Unicode 처리
plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
```

```python
df = pd.read_csv("../../data/student.csv", encoding='utf-8')
df
```

```python
fig = plt.figure()
```

### subplots(col, row) : 행x열 의 그리드를 갖는 차트 그룹 생성

```python
fig, axes = plt.subplots(1,3)
```

```python
fig # 그래프의 형태를 갖고 있음
```

```python
axes # 실제 그래프 데이터를 입력 ( 배열 형태이기 떄문에 인덱스 조회를 통해 값 부여 )
```

***

```python
fig, axes = plt.subplots(2,2, figsize=(15,10))
fig.suptitle('여러 개의 그래프')
##############################################################
axes[0, 0].bar('이름', '국어', data=df, label='국어점수')
axes[0, 0].set_title('학생별 국어 점수')
axes[0, 0].legend()
axes[0, 0].set_xlabel('이름')
axes[0, 0].set_ylabel('점수')
axes[0, 0].grid(linestyle='--', linewidth=0.5)
axes[0, 0].set_facecolor('lightyellow')

##############################################################
axes[0, 1].plot('이름', '영어', data=df, label='영어점수')
axes[0, 1].plot('이름', '수학', data=df, label='수학점수')
axes[0, 1].set_title('학생별 영어/수학 점수')
axes[0, 1].legend()
axes[0, 1].set_xlabel('이름')
axes[0, 1].set_ylabel('점수')

##############################################################
axes[1, 0].barh('이름', '키', data=df)
axes[1, 0].set_title('학생별 키')
axes[1, 0].set_xlabel('키')
axes[1, 0].set_ylabel('이름')

##############################################################
axes[1, 1].plot('이름', '사회', data=df, color='g')

##############################################################
```
