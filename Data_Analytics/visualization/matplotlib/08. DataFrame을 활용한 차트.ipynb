```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['axes.unicode_minus'] = False # " - " 기호 Unicode 처리
plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
```

```python
data = {
    "이름" : ["채치수", "정대만", "송태섭", "서태웅", "강백호", "변덕규", "황태산", "윤대협"],
    "학교" : ["북산고", "북산고", "북산고", "북산고", "북산고", "능남고", "능남고", "능남고"],
    "키" : [197, 184, 168, 187, 188, 202, 188, 190],
    "국어" : [90, 40, 80, 40, 15, 80, 55, 100],
    "영어" : [85, 35, 75, 70, 20, 100, 65, 85],
    "수학" : [100, 50, 70, 70, 10, 85, 45, 90],
    "과학" : [95, 55, 80, 75, 35, 85, 40, 95],
    "사회" : [85, 25, 75, 80, 10, 80, 35, 95],
    "SW특기" : ["Python", "Java", "Javascript", np.nan, np.nan, "C", "PYTHON", "C#"]
}

df = pd.DataFrame(data)
df
```

```python
plt.plot(df['이름'], df['국어'])
```

```python
plt.plot('이름', '국어', data=df)
```

```python
plt.barh('이름', '영어', data=df)
```

```python
plt.barh('이름', '영어', data=df.sort_values('영어'))
```

```python
plt.figure(figsize=(10, 5))

plt.plot('이름', '국어', data=df, label='국어')
plt.plot('이름', '영어', data=df, label='영어')
plt.plot('이름', '수학', data=df, label='수학')
plt.legend(ncol=3)
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.plot('이름', '국어', data=df, label='국어')
plt.plot('이름', '영어', data=df, label='영어')
plt.plot('이름', '수학', data=df, label='수학')
plt.legend(ncol=3)
plt.grid(axis='y', alpha=0.3, color='r', ls='--')

plt.show()
```
