## matplotlib (https://matplotlib.org/stable/index)

- 다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['axes.unicode_minus'] = False # " - " 기호 Unicode 처리
plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
plt.rcParams['font.size'] = 15 # 폰트 사이즈
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x, y)
```

```python
days = [1,2,3]
az = [2,4,8]
pfizer = [5,1,3]
moderna = [1,2,5]
```

```python
plt.plot(days, az)
plt.plot(days, pfizer)
plt.plot(days, moderna)
```

```python
plt.plot(days, az, label="az", marker='o', ls='--')
plt.plot(days, pfizer, label="pfizer", marker='s', ls='-.')
plt.plot(days, moderna, label="moderna")
plt.legend(ncol=3)
```
