```python
import seaborn as sns

sns.set(font_scale=1) # seaborn 기초 세팅
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['axes.unicode_minus'] = False # " - " 기호 Unicode 처리
plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
```

```python
df = sns.load_dataset('flights')
df.head()
```

```python
df.shape
```

```python
sns.lineplot(data=df, x='year', y='passengers')
```

```python
sns.lineplot(data=df, x='year', y='passengers')
plt.xticks(df['year'], rotation=45);
```

#### 1) hue : 그룹화 

```python
sns.lineplot(data=df, x='year', y='passengers', hue='month') # hue : 그룹화
```

#### 2) palette
> https://seaborn.pydata.org/tutorial/color_palettes.html|

```python
sns.lineplot(data=df, x='year', y='passengers', hue='month', palette='Set2')
```

#### 3) style

```python
sns.lineplot(data=df, x='year', y='passengers', hue='month', palette='Set2', style='month')
```

#### 4) marker

```python
sns.lineplot(data=df, x='year', y='passengers', hue='month', palette='Set2', style='month', markers=True)
```

#### 5) err_style : 오차범위 구간표시 스타일

```python
sns.lineplot(data=df, x='year', y='passengers', err_style='bars')
```
