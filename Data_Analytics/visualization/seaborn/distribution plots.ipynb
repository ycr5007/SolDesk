```python
import seaborn as sns

sns.set(font_scale=1) # seaborn 기초 세팅
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['axes.unicode_minus'] = False # " - " 기호 Unicode 처리
plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
```

```python
iris_df = sns.load_dataset('iris')
iris_df.head()
```

#### 1) histplot

```python
sns.histplot(iris_df['sepal_length'])
```

```python
sns.histplot(iris_df['sepal_length'], bins=20)
```

```python
sns.histplot(iris_df['sepal_length'], bins=20, kde=True) # kde : 커널밀도, 확률밀도
```

```python
sns.histplot(iris_df['sepal_length'], bins=20, kde=True, stat='density')
```

#### Group

```python
plt.figure(figsize=(15,6))
sns.histplot(data=iris_df, x='sepal_length', bins=20, hue='species', kde=True, stat='density')
# ValueError: The following variable cannot be assigned with wide-form data: `hue` :: 데이터를 시리즈로 불러왔기 때문에 에러 발생
```

```python
sns.histplot(data=iris_df, x='sepal_width', y='sepal_length')
```

#### 2) distplot

```python
sns.distplot(iris_df['sepal_width'])
```

#### 3) jointplot

```python
sns.jointplot(data=iris_df, x='sepal_width', y='sepal_length')
```
