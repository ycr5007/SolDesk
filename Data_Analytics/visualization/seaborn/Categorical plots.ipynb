#### 6) countplot

```python
titanic_df = sns.load_dataset('titanic')
titanic_df.head()
```

```python
titanic_df.isnull().sum()
```

###### 1) 성별차트 - pandas

```python
titanic_df.groupby('sex')['sex'].count().plot.bar(rot=0, color=['red', 'blue'])
```

#### 2) 성별차트 - seaborn

```python
sns.countplot(data=titanic_df, x='sex')
```

```python
sns.countplot(data=titanic_df, y='sex')
```

#### 성별인원수를 객실 등급별로 시각화

```python
titanic_df.groupby(['sex', 'class'])['sex'].count().plot.bar(rot=45)
```

```python
sns.countplot(data=titanic_df, x='sex', hue='class', palette='flare')
```
