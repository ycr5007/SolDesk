```python
import seaborn as sns

sns.set(font_scale=1) # seaborn 기초 세팅
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['axes.unicode_minus'] = False # " - " 기호 Unicode 처리
plt.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
```

```python
iris_df = sns.load_dataset('iris')
```

```python
sns.scatterplot(data=iris_df, x='petal_length', y='petal_width')
```

```python
sns.scatterplot(data=iris_df, x='sepal_length', y='sepal_width')
```

```python
sns.scatterplot(data=iris_df, x='sepal_length', y='sepal_width', hue='species')
```

#### petal_length 값을 3 미만 , 5미만, 5이상으로 분류하여 표시
- petal_length2 컬럼을 생성해서 petal_length 값을 3미만 s, 5미만 m, 5이상 l 으로 분류하여 표시

```python
iris_df['petal_length'].unique()
```

```python
def function1(x):
    if x < 3:
        return 's'
    elif x < 5:
        return 'm'
    else:
        return 'l'
```

```python
iris_df['petal_length2'] = iris_df['petal_length'].apply(function1)
```

```python
iris_df.head()
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(data=iris_df, x='sepal_length', y='sepal_width', hue='species', style='petal_length2')
```

#### petal_length 값을 3 미만 , 5미만, 5이상으로 분류하여 표시
- petal_length3 컬럼을 생성해서 petal_length 값을 3미만 3, 5미만 5, 5이상 7 으로 분류하여 표시

```python
def function2(x):
    if x < 3:
        return 3
    elif x < 5:
        return 5
    else:
        return 7
```

```python
iris_df['petal_length3'] = iris_df['petal_length'].apply(function2)
iris_df.head()
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(data=iris_df, x='sepal_length', y='sepal_width', hue='species', size='petal_length3')
```

### replot : col 기준으로 따로 그려준다 

```python
sns.relplot(data=iris_df, x='sepal_length', y='sepal_width', col='species')
```

### lmplot : 산점도 + 회귀곡선

```python
sns.lmplot(data=iris_df, x='sepal_length', y='sepal_width', hue='species')
```
