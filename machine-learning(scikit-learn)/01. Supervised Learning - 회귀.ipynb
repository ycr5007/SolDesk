## 2. 선형회귀
### 1) 단순선형회귀(Simple Linear Regression)


- <b style="color:red">독립변수가 1 개인 경우 사용</b>
- <b style="color:red">X,Y는 선형관계</b>






- 입력 변수가 X, 출력변수가 Y 일 때, 단순선형회귀의 회귀식은 검은 선으로 나타낼 수 있음
- 잔차(오차) : 정규분포 추종, Y : 정규분포 추종

- 모집단의 전체 데이터를 가지고 작업을 할 수는 없기 때문에 모집단을 대표하는 데이터로 추정해서 사용

### - 회귀 계수 추정



- 잔차의 제곱합(SSE) 를 하는 이유?
    - 잔차의 합이 0 이 되는 해는 무수히 많기 때문에
    - 미분이 가능해야 함

### - 회귀 계수 해석



- 베타1 이 0.203, 기울기 9.312
- p-value가 낮아지면 유의성이 높음

### - 선형 회귀의 정확도 평가

- Y의 변동성을 독립변수가 얼마나 잘 설명하느냐가 중요함
- 변수가 여러 개 일때 각각 Y를 설명하는 변동성이 크면 좋은 변수(Y의 값을 설명할 수 있는 부분이 크면 좋다는 의미)

### 2) 다중선형회귀(Simple Linear Regression)

- <b style="color:red">독립변수가 여러개인 경우 사용</b>
- 단순선형회귀 분석과 동일하게 SSE를 최소화 하는 방향으로 추정



- 다른 변수(radio, newspaper)가 고정되어 있을 때 TV가 1 단위 증가할 때 TV로 인한 매출이 0.046 증가 => 대립가설 채택
- 다른 변수(tv, radio)가 고정되어 있을 때 newspaper 1 단위 증가할 때 newspaper로 인한 매출이 -0.001 감소한 것처럼 나왔으며 p-value 가 커지게 됨 => 귀무 가설 채택

#### - 회귀 계수 검정
    - 귀무가설 : 회귀계수는 0 이다(변수의 설명력이 하나도 존재하지 않음)
    - 대립가설 : 회귀계수는 0이 아니다(설명력이 있는 변수가 존재한다)
    - T 검정

- <b style="color:red">모델 검정</b>
    - 귀무가설 : 모든 회귀계수는 0 이다(변수의 설명력이 하나도 존재하지 않음)
    - 대립가설 : 하나의 회귀계수라도 0이 아니다(설명력이 있는 변수가 존재한다)
    - F검정

#### - 다중 공선성(Multicollinearity)
- 독립변수들이 강한 선형 관계에 있을 때 다중공선성이 있다고 함



- radio, newspaper 하나의 독립변수가 매출에 끼치는 영향을 알아 봤을 때, p-value 가 0.0001 보다 작기 때문에 두 변수가 유의미 하다고 판단함
- TV, radio, newspaper 세 개의 변수가 추가되자 newspaper의 회귀 계수가 거의 0에 가까워졌고, p-value 값이 높아짐(다중공선성 때문에 발생)

- 변수가 늘어나게 되면 Y 값을 설명하기 위한 변수들이 증가하는 것
- Y 크기는 정해져 있고, 변수의 갯수가 늘어나기 때문에 특정 변수는 다른 변수와 겹치는 부분이 있고, 중복으로 가져가지 못함
- 이러면서 회귀 계수는 낮아지고 p-value 는 높아짐
- <b>잘못된 변수 해석, 예측 정확도 하락, 모델을 만드는 데 시간이 많이 걸림</b>

#### - 다중 공선성 해결
- Feature Selection : 중요 변수만 선택하는 방법
    - 단순히 변수를 제거하는 방법(correlation(상관) 등의 지표를 보고)
    - Lasso
    - Stepwise
    - 기타 변수 선택 알고리즘(유전 알고리즘 등)

- 변수를 줄이지 않고 활용하는 방법
    - AutoEncoder 등의 Feature Extraction 기법(딥러닝 기법)
    - PCA
    - Ridge
