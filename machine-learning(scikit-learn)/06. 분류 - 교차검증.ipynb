## 교차 검증

- 알고리즘을 학습시키는 학습 데이터와 이에 대한 예측 성능을 평가하기 위한 별도의 테스트 용 데이터 필요
    - 과적합(모델이 학습 데이터에만 과도하게 최적화 됨) 발생 가능성이 있음
<br>
- 데이터 편중을 막기 위해서 별도의 여러 세트로 구성된 학습 데이터 세트와 검증 데이터 세트에서 학습과 평가를 수행
- 수행 결과에 따라 하이퍼 파라미터 튜닝 등의 모델 최적화 가능

- 학습 데이터를 다시 분할하여 학습 데이터와 학습된 모델의 성능을 일차 평가하는 검증 데이터로 나눔
- 테스트 데이터 셋 : 모든 학습/검증 과정이 완료된 후 최종적으로 성능을 평가하기 위한 데이터 세트

- K 폴드의 문제점 : 위에서도 보면 인덱스가 연속된 형태로 잘라서 작업을 해 버림(값의 분포가 한 방향으로 치우치게 됨
    - 대출 사기 데이터를 예측하는 알고리즘을 구현한다 할 때, 전체 데이터 셋이 1억건이고, 대출 사기가 약 1000 건이라 할 때 0.0001% 의 아주 작은 확률로 대출 사기 레이블이 존재하는 것 => 만일 데이터 셋을 나눴는데 특정 데이터 셋에 사기 데이터가 몰려 있다면 문제임
- 원본 데이터의 레이블 분포를 고려한 뒤 이 분포와 동일하게 학습 데이터와 검증 데이터 셋 분배 해 줌
