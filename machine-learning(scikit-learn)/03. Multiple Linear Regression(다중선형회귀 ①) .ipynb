### 원-핫 인코딩

- 공부장소의 경우 현재 데이터가 범주형(문자) 데이터로 들어가 있음
- 범주형 데이터를 표현할 때 원-핫 인코딩으로 처리
- 표현하고 싶은 값만 1로 처리하고 나머지는 0으로






#### Multicollinearity(다중 공선성)
- 독립변수들 간에 서로 강한 상관관계를 가지면서 회귀계수 추정의 오류가 나타나는 문제
- 하나의 피처가 다른 피처에 영향을 끼치는 상황
- Dummy variable trap 이라고도 함
- <b>해결방법 : Dummy Column 이 n 개인 경우 n-1 개만 사용</b>




### 다양한 평가 지표 (회귀 모델)

<b>1. MAE (Mean Absolute Error) : (실제 값과 예측 값) 차이의 절대값</b>



- 실제값과 예측값 사이에는 12.8 점 정도의 차이가 난다

**2. MSE (Mean Squared Error) : 차이의 제곱**



- 1 미만의 값은 더 작아지고, 1 이상의 값은 더 커지는 속성이 있기 때문에 이상값에 대해서 민감하게 반응이 된다
- 실제 데이터보다 커지기 때문에(제곱하니까) 데이터를 해석하기 위해 다른 부분을 고려해야 한다.
    - 100점 만점인데 176점이나 차이가 난다는 건 말이 안됨

**3. RMSE (Root Mean Squared Error) : 차이의 제곱에 루트**


- 작은 값은 더 작아지고 큰 값은 더 커지는 부분 해결
- 원본 데이터랑 같은 단위기 때문에 위의 경우랑 다르게 해석이 편리해짐

**4. R2 : 결정 계수**
