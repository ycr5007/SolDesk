## 지도학습의 종류
> ### 회귀(Regression)와 분류(Classification)

> ### Regression
> - 변수들 간의 상관관계를 찾는 것, 연속적인 데이터로부터 결과를 예측
> - 예측 결과가 숫자일 때 주로 사용
> - ex) 근속연수에 따른 임금, 키에 따른 몸무게, 사용 기간에 따른 스마트폰 가격...

> ### Classification
> - 주어진 데이터를 정해진 범주(category)에 따라 분류
> - 예측 결과가 숫자가 아닐 때 주로 사용
> - ex) 스팸 메일 필터링, 시험 합격 여부, 재활용 분리수거 품목, 악성 종양 여부...
