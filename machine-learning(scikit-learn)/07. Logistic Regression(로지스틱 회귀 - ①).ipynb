# 4. Logistic Regression(로지스틱 회귀)

- 선형회귀방식을 분류에 적용한 알고리즘
- Sigmoid(시그모이드) 함수 최적선을 찾고 이 sigmoid 함수의 반환 값을 확률로 간주해 확률로 분류를 결정함


- p 는 항상 최소 0, 최대 1 사이의 값만 가질 수 있음

### 공부 시간에 따른 자격증 시험 합격 가능성



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("./dataset/LogisticRegressionData.csv")
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### 로지스틱 회귀 모델
- sigmoid function 

```python
from sklearn.linear_model import LogisticRegression

lr_clf = LogisticRegression()
```

```python
lr_clf.fit(X_train, y_train)
```

```python
lr_clf.predict([[6]])
```

```python
lr_clf.predict_proba([[6]])
```

```python
lr_clf.predict([[4]])
```

```python
lr_clf.predict_proba([[4]])
```

```python
pred_y = lr_clf.predict(X_test)
pred_y
```

```python
X_test
```

```python
lr_clf.score(X_test, y_test)
```

### 시각화

```python
# X 축 데이터 증가

X_range = np.arange(min(X), max(X), 0.1)
X_range
```

#### 식 적용 


```python
p = 1 / ( 1 + np.exp(-(lr_clf.coef_ * X_range + lr_clf.intercept_)))
p
```

```python
X_range.shape
```

```python
# 2 차원 -> 1 차원
p = p.reshape(-1)
p
```

```python
p.shape
```

```python
plt.scatter(X_train, y_train, color="blue")
plt.plot(X_range, p, color="green")
plt.plot(X_range, np.full(len(X_range), 0.5), color="red")
plt.title("Probability by hours")
plt.xlabel("hours")
plt.ylabel("P")
plt.show()
```

```python
lr_clf.predict_proba([[4.5]])
```

### 혼동 행렬 (Confusion Matrix)
- classfication 형태로 나타내는 모델의 성능 지표





```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, pred_y)
```
